```python
import pandas as pd
import numpy as np
np.random.seed(42)

data = {
    "Order_ID": np.arange(1001, 1021),
    "Product": np.random.choice(["Laptop", "Mobile", "Tablet", "Headphones"], 20),
    "Category": np.random.choice(["Electronics", "Accessories"], 20),
    "Quantity": np.random.randint(1, 5, 20),
    "Price_per_Unit": np.random.randint(500, 50000, 20),
    "Region": np.random.choice(["North", "South", "East", "West"], 20),
    "Order_Date": pd.date_range(start="2025-01-01", periods=20)
}

df = pd.DataFrame(data)
df["Total_Sales"] = df["Quantity"] * df["Price_per_Unit"]

print(df)
df.to_csv("sales_dataset.csv", index=False)
```
